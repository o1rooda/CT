## 1. k-nn(k-nearest neighbor k-최근접이웃)
<li>기존 데이터 중 가장 유사한 k개의 데이터를 이용해서 값을 예측하는 방법</li>
<li>knn은 매우 단순하지만 정확성이 비교적 높음</li>

### algorithm knn
<li>input : inX(테스트데이터), dataSet(학습데이터), labels(class), k(투표갯수)</li>
<li>inX : training dataset m(n_features) x n(n_samples)</li>
<li>dataSet : test instance</li>
<li>labels : class label</li>
<li>k : the number of most similar examples in S</li>
<li>output : class of x majority class</li>

### 알고리듬에 따른 개발

<li> 1. data read - 데이터 준비</li>
<li> 2. distance - 학습데이터와 테스트데이터 거리(유사도) 계산</li>
<li> 3. select k neighbors - 가장 가까운 k개 선정</li>
<li> 4. decision - k개에서 voting하여 결정</li>
<li> 5. accuracy - 정확성</li>

```python
import os
import operator
import pylab
import numpy as np
from numpy import *

def createDataSet():
    group = array([[1.0,1.1],[1.0,1.0],[0,0],[0,0.1]])
    labels = ['A','A','B','B']
    return group, labels

def classify0(inX, dataSet, labels, k):
    dataSetSize = dataSet.shape[0]
    diffMat = tile(inX, (dataSetSize,1)) - dataSet
    sqDiffMat = diffMat**2
    sqDistances = sqDiffMat.sum(axis=1)
    distances = sqDistances**0.5
    sortedDistIndicies = distances.argsort()
    classCount={}
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel,0) + 1
    sortedClassCount = sorted(classCount.iteritems(), key=operator.itemgetter(1), reverse=True) #아이템 정렬
    return sortedClassCount[0][0]
```

## 2. SVM(Support vector Machine)
<li>특정 공간에서 주어진 두 분류의 데이터를 구분지을 수 있는 최적(optimal)의 초평면(hyperplane)</li>
<li>support vector를 찾는 알고리즘으로 두 분류에서 가장 가까운 데이터를 하나씩 찾아서 그 거리를 계산했을 때 가장 멀어질 수 있는 초평면을 찾는 것을 목표</li>

#### ** hyperplane : 기하학에서 초평면은 주변의 공간보다 낮은 한 차원의 부분 공간이다. 만약 2차원 공간이면 초평면은 1차원 선. 이 개념은 부분 공간의 차원의 컨셉이 정의된 모든 일반적인 공간 내에서 사용된다.

<li>feature들은 그림과 같이 공간에 vector로 표시</li>
<li>하얀 색 vector들을 A그룹에 속하는 white point, 그 반대로 검은색 vector들을 B그룹에 속하는 black point</li>
<li>벡터들 중 가장 가까운 벡터와 수직거리로 가장 먼 거리를 가진 hyperplane이 2그룹을 효과적으로 분류</li>
<li>이때 가장 가까운 벡터들을 support vector</li>
<li>두 그룹을 구분하기 위해서 Maximum Margin Hyperplane 찾기</li>
#### ** Margin : 초평면 분리에 가장 가까운 지점을 찾기를 원하면서, 이 지점이 가능한 한 분리선에서 멀리 떨어져 있는 것.
